```python
data_2['Monthly_cash_advance'] = data_2['CASH_ADVANCE']/data_2['TENURE']
```

#### b)Purchases by type (one-off, installments):

```python
def pbt(data_2):
    if((data_2['ONEOFF_PURCHASES']==0)&(data_2['INSTALLMENTS_PURCHASES']==0)):
        return 'None'
    elif((data_2['ONEOFF_PURCHASES']>0)&(data_2['INSTALLMENTS_PURCHASES']==0)):
        return 'One_Of' 
    elif((data_2['ONEOFF_PURCHASES']==0)&(data_2['INSTALLMENTS_PURCHASES']>0)):
        return 'Installment_Purchases'
    else:
        return 'Both'
```

```python
data_2['purchases_by_types'] = data_2.apply(pbt,axis=1)
```

#### Choosing no. of clusters using Silhouette Coefficient:

```python
from sklearn import metrics
```

```python
metrics.silhouette_score(data_scaled_2, km_7.labels_)
```

```python
k_range = range(3,12)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=123)
    km.fit(data_scaled_2)
    scores.append(metrics.silhouette_score(data_scaled_2,km.labels_))
```

```python
scores
```

```python
plt.plot(k_range, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```

 If we take highest SC score, 5 segment solution is best

```python
data_updated.cluster_5.value_counts()/sum(data_updated.cluster_5.value_counts())
```

#### Elbow Analysis:

```python
cluster_range = range( 1, 20 )
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans( num_clusters )
    clusters.fit( data_scaled_2 )
    cluster_errors.append( clusters.inertia_ )
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )

clusters_df.head(10)
```

```python
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

The elbow diagram shows that the gain in explained variance reduces significantly from 3 to 4 to 5 to 6. So, optimal number of clusters could either 4 or 5 or 6,because 3 cluster size have un-even distribution.

### 6.Strategic insights and implementation of strategies:

- From Elbow Analysis we found out that either 3 or 4 or 5 or 6 would be a good cluster size.
- But since 3 cluster size had un-even distribution we rejected it.
<br>
<br>
- For 4 cluster size, we following groups: 
    - Low balance,installment type	
    - Active customer with good cash	
    - Inactive customer with low balance	
    - high cash but no actives
<br>
<br>
- For 5 cluster size, we following groups: 
    - Loan guy with good purchase frequency	
    - Low balance,installment type	
    - Active customer with good cash	
    - Inactive customer with low balance	
    - high cash but no actives     
<br>
<br>    
- For 6 cluster size, we following groups: 
    - One_off with no loan type
    - Loan guy with good purchase frequency	
    - Low balance,installment type	
    - Active customer with good cash	
    - Inactive customer with low balance	
    - high cash but no actives      

Strategies for different groups:-
- <strong><em>One_off with no loan type</em></strong>               
Solution=>The customer most of the times,purchase the product in one-go rather then going for installments.So,<strong>flat discount,cashback,different offers on the card usage</strong> would interest the customer.
<br>
<br>
- <strong><em>Loan guy with good purchase frequency</em></strong>	
Solution=> He/She is interested in buying stuff and takes small loans to do that.So, <strong>'different loan plans'</strong> would definitely attract them.
<br>
<br>
- <strong><em>Low balance,installment type</em></strong>	
Solution=><strong>'Different EMI's' offers</strong> would benefit here.
<br>
<br>
- <strong><em>Active customer with good cash</em></strong>
<br>Solution=><strong>Limited edition cards,special offers,etc</strong> would be enough.
<br>
<br>
- <strong><em>Inactive customer with low balance</em></strong>
<br>Solution=><strong>'Cash-advance'</strong> options would be good here as a strategy.
<br>
<br>
- <strong><em>high cash but no actives</em></strong>               
Solution=>The person have good amount of money but don't do that much of activities.So,<strong>'attract card offers'</strong> like cash-back,discount on usage,etc would be helpful.
